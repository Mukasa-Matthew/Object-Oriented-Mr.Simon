```python

class Person:
    def __init__(self, name):
        self.name = name


class Student(Person):
    def __init__(self, name, school):
        super().__init__(name)
        self.school = school


class Boy(Student):
    def __init__(self, name, school, hobby):
        super().__init__(name, school)
        self.hobby = hobby


class Girl(Student):
    def __init__(self, name, school, interests):
        super().__init__(name, school)
        self.interests = interests

    def send_message(self, recipient_school):
        print(f"{self.name} sends a secret message to a boy at '{recipient_school}'.")

    def pick_friend(self, boy1, boy2):
        print("\nNow she must decide who to befriend based on shared hobbies...")
        if boy1.hobby in self.interests:
            print(f"{self.name} chooses {boy1.name} because they both love '{boy1.hobby}'.")
        elif boy2.hobby in self.interests:
            print(f"{self.name} chooses {boy2.name} because they both enjoy '{boy2.hobby}'.")
        else:
            print(f"{self.name} stays friends with both — she can’t decide!")


# Story Simulation
def story():
    # Schools
    correct_school = "Trinity High"
    mistaken_school = "Trinity Vocational"

    # Two boys
    boy1 = Boy("Brian", correct_school, "painting")
    boy2 = Boy("Allan", mistaken_school, "cycling")

    # The girl
    girl = Girl("Lydia", "St. Catherine Girls", ["reading", "cycling"])

    # The mix-up
    print(f"Lydia wants to message {boy1.name} at {boy1.school}, but she accidentally sends it to '{mistaken_school}'.\n")
    girl.send_message(mistaken_school)

    print(f"\nThe message reaches {boy2.name} at {boy2.school}, who replies kindly.")
    print(f"{boy2.name}'s funny replies make Lydia curious and soon they start chatting often.\n")

    print("When she learns about the mistake, she has to decide who she really wants to stay close to.")

    # Decision time
    girl.pick_friend(boy1, boy2)



if __name__ == "__main__":
    story()
```
